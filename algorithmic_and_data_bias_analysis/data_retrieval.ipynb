```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("compas.db")
```

```python
people = pd.read_sql_query("SELECT * from people", con)
people.head()
```

```python
jail = pd.read_sql_query("SELECT * from jailhistory", con)
jail.head()
```

```python
prison = pd.read_sql_query("SELECT * from prisonhistory", con)
prison.head()
```

```python
compas = pd.read_sql_query("SELECT * from compas", con)
compas.head()
```

```python
charge = pd.read_sql_query("SELECT * from charge", con)
charge.head()
```

```python
casearrest = pd.read_sql_query("SELECT * from casearrest", con)
casearrest.head()
```

```python
data = people.merge(jail.drop(columns=['id', 'first', 'last', 'dob']), how='left', left_on='id', right_on='person_id')
data = data.merge(prison.drop(columns=['id', 'name', 'first', 'middle', 'last', 'dob']), 
                  how='left', left_on='id', right_on='person_id')
data = data.merge(compas.drop(columns=['id', 'first', 'last', 'compas_person_id', 'compas_case_id', 
                                       'compas_assessment_id', ]), how='left', left_on='id', right_on='person_id')
#data = data.merge(charge.drop(columns=['id']), how='left', left_on='id', right_on='person_id')
#data = data.merge(casearrest.drop(columns='id'), how='left', left_on='id', right_on='person_id')
```

```python
data.to_csv('compas.csv')
```
